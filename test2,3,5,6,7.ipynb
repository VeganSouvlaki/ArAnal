```python
import numpy as np

# Συνάρτηση που ελέγχει αν όλα τα στοιχεία ενός διανύσματος είναι μικρότερα από ένα όριο ανοχής
# Χρησιμοποιείται για να ελέγξει αν η μέθοδος έχει συγκλίνει
def check(x, tol):
    return np.all(abs(x) < np.ones(len(x)) * tol)

# Υλοποίηση της μεθόδου Gauss-Jacobi για την επίλυση του συστήματος Ax = b
# Η μέθοδος ενημερώνει όλες τις τιμές του x στο τέλος κάθε επανάληψης (ταυτόχρονη ενημέρωση)
def gauss_jacobi(a, b, x, tol=1e-7):
    n = len(b)
    x.fill(0.0)  # Αρχικοποιούμε το x με μηδενικά (πρώτη προσέγγιση)
    iterations = 0  # Μετρητής επαναλήψεων

    while True:
        y = b - np.matmul(a, x)  # Υπολογίζουμε το y = b - Ax (υπόλοιπο)

        for i in range(n):
            y[i] /= a[i, i]  # Διαιρούμε κάθε y[i] με το αντίστοιχο διαγώνιο στοιχείο a[i,i]

        x += y  # Προσθέτουμε το διάνυσμα διόρθωσης y στο x
        iterations += 1

        if check(y, tol):  # Αν οι διορθώσεις είναι πολύ μικρές, έχουμε συγκλίνει
            break

    return x, iterations  # Επιστρέφουμε τη λύση και τον αριθμό επαναλήψεων

# Υλοποίηση της μεθόδου Gauss-Seidel για την επίλυση του συστήματος Ax = b
# Η μέθοδος ενημερώνει τις τιμές του x μόλις υπολογιστούν, μέσα στην ίδια επανάληψη
# Έτσι χρησιμοποιεί τις "φρεσκότερες" τιμές και συνήθως συγκλίνει πιο γρήγορα

def gauss_seidel(a, b, x, tol=1e-7):
    n = len(b)
    x.fill(0.0)  # Αρχικοποιούμε το x με μηδενικά
    iterations = 0  # Μετρητής επαναλήψεων

    while True:
        y = b - np.matmul(a, x)  # Υπολογισμός του υπολοίπου

        for i in range(n):
            y[i] /= a[i, i]  # Κανονικοποίηση ως προς το διαγώνιο στοιχείο
            x[i] += y[i]     # Ενημέρωση του x[i] αμέσως (χρήση εντός της ίδιας επανάληψης)

        iterations += 1

        if check(y, tol):  # Έλεγχος σύγκλισης
            break

    return x, iterations

# Κύρια συνάρτηση εκτέλεσης του προγράμματος
def main():
    n = 4  # Διαστάσεις συστήματος 4x4
    tol = 1e-7  # Όριο ανοχής για σύγκλιση

    # Ορισμός πίνακα A (συντελεστές του συστήματος)
    a = np.array([
        [12.1, 3.9, 0.3, -4.1],
        [4.3, -11.3, 0.8, 1.5],
        [1.0, -2.8, 14.3, -8.1],
        [2.4, 6.1, -1.1, 12.5]
    ])

    # Ορισμός διανύσματος b (δεξιά μέλη)
    b = np.array([1.2, 2.3, 3.4, 4.5])
    x = np.zeros(n)  # Αρχικοποίηση του διανύσματος x

    # Εμφάνιση του συστήματος προς επίλυση
    print("Πίνακας A:\n", a)
    print("\nΔιάνυσμα b:\n", b)
    print("\nΛύνουμε το σύστημα A·x = b\n")

    # Επίλυση με Gauss-Jacobi
    x_jacobi, iter_jacobi = gauss_jacobi(a, b, x.copy(), tol)
    print("Λύση με Jacobi:")
    for i in range(n):
        print(f"x[{i}] = {x_jacobi[i]}")
    print(f"Αριθμός επαναλήψεων: {iter_jacobi}\n")

    # Επίλυση με Gauss-Seidel
    x_seidel, iter_seidel = gauss_seidel(a, b, x.copy(), tol)
    print("Λύση με Seidel:")
    for i in range(n):
        print(f"x[{i}] = {x_seidel[i]}")
    print(f"Αριθμός επαναλήψεων: {iter_seidel}\n")

# Εκκίνηση προγράμματος
main()
```

```python
import numpy as np

# Εκτύπωση πίνακα σε μορφή πίνακα nxn
def print_matrix(a):
    n = a.shape[0]
    for i in range(n):
        for j in range(n):
            print(f"{a[i,j]:.6g}", end=" ")
        print()
    print()

# Αντικατάσταση τιμών για την κανονικοποίηση των γραμμών
def normalize(a, b, k):
    n = a.shape[0]
    for j in range(k, n):
        maxv = max(abs(a[j, k:]).max(), abs(b[j, :]).max())
        if maxv != 0:
            a[j, k:] /= maxv
            b[j, :] /= maxv

# Εναλλαγή τιμών δύο μεταβλητών
def swap(a, b):
    return b, a

# Εναλλαγή γραμμών στον πίνακα a και b
def swapa(a, b, k, p):
    a[[k, p], k:] = a[[p, k], k:]
    b[[k, p], :] = b[[p, k], :]

# Επιλογή κύριου στοιχείου για pivoting
def pivot(a, b, k):
    normalize(a, b, k)
    n = a.shape[0]
    maxp = k
    for p in range(k+1, n):
        if abs(a[p, k]) > abs(a[maxp, k]):
            maxp = p
    if maxp != k:
        swapa(a, b, k, maxp)

# Συνάρτηση για αντιστροφή πίνακα με Gaussian elimination με pivoting
def invert(a):
    n = a.shape[0]
    a = a.copy()
    b = np.identity(n)

    for k in range(n-1):
        pivot(a, b, k)
        for i in range(k+1, n):
            g = -a[i, k] / a[k, k]
            a[i, k:] += a[k, k:] * g
            b[i, :] += b[k, :] * g

    x = np.zeros((n, n))
    for i in reversed(range(n)):
        for j in range(n):
            x[i, j] = b[i, j] - np.dot(a[i, i+1:], x[i+1:, j])
            x[i, j] /= a[i, i]

    return x

# Κύριο πρόγραμμα
def main():
    a = np.array([
        [2.1, 3.9, 0.3, -4.1],
        [4.3, -1.3, 0.8, 1.5],
        [1.0, -2.8, 4.3, -8.1],
        [2.4, 6.1, -1.1, 12.5]
    ])

    print("Αρχικός πίνακας A:")
    print_matrix(a)

    ainv = invert(a)

    print("Αντίστροφος πίνακας A^(-1):")
    print_matrix(ainv)

    identity_approx = np.matmul(a, ainv)
    print("Έλεγχος: A · A^(-1) = Ι:")
    print_matrix(identity_approx)

main()
```

```python
import numpy as np

# ---------------- Βοηθητικές Συναρτήσεις ---------------- #

# Κανονικοποιεί τις γραμμές από τη γραμμή k και κάτω, για να βελτιώσει τη σταθερότητα
def normalize(a, b, k):
    n = len(b)
    for j in range(k, n):
        maxv = max(abs(b[j]), max(abs(a[j, k:])))
        a[j, k:] /= maxv
        b[j] /= maxv

# Ανταλλάσσει τις γραμμές k και p σε a και b
def swap(a, b, k, p):
    a[[k, p]] = a[[p, k]]
    b[k], b[p] = b[p], b[k]

# Επιλέγει τη γραμμή με το μεγαλύτερο pivot στη στήλη k και την φέρνει στην k-οστή θέση
def pivot(a, b, k):
    normalize(a, b, k)
    maxp = k + np.argmax(np.abs(a[k:, k]))
    if maxp != k:
        swap(a, b, k, maxp)

# Τριγωνοποιεί τον πίνακα a και τροποποιεί αντίστοιχα το b (απαλοιφή Gauss)
def triang(a, b):
    n = len(b)
    for k in range(n - 1):
        pivot(a, b, k)
        for i in range(k + 1, n):
            factor = a[i, k] / a[k, k]
            a[i, k:] -= factor * a[k, k:]
            b[i] -= factor * b[k]

# Οπισθοδρόμηση για επίλυση του τριγωνοποιημένου συστήματος
def backsub(a, b):
    n = len(b)
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - np.dot(a[i, i + 1:], x[i + 1:])) / a[i, i]
    return x

# Συνδυασμός των παραπάνω για επίλυση γραμμικού συστήματος
def gauss_solve(a, b):
    a = a.copy()
    b = b.copy()
    triang(a, b)
    return backsub(a, b)

# ---------------- Μη Γραμμικό Σύστημα ---------------- #

# Υπολογίζει τις τιμές των εξισώσεων f(x) και τον Jacobian πίνακα
def nonlinear_system(x):
    f = np.zeros(3)
    J = np.zeros((3, 3))

    # f(x)
    f[0] = x[0] + x[1] + x[2] - 3.0
    f[1] = x[0]**2 * x[1] + x[1]**2 * x[2] + x[2]**2 * x[0] - 4.0
    f[2] = x[0]**2 + x[1]**2 + x[2]**2 - 5.0

    # Jacobian
    J[0, :] = [1.0, 1.0, 1.0]
    J[1, 0] = 2 * x[0] * x[1] + x[2]**2
    J[1, 1] = 2 * x[1] * x[2] + x[0]**2
    J[1, 2] = 2 * x[2] * x[0] + x[1]**2
    J[2, :] = 2 * x

    return f, J

# ---------------- Κύρια Συνάρτηση ---------------- #

def main():
    tol = 1e-7
    x = np.array([1.0, 1.5, 2.1])  # αρχική προσέγγιση

    while True:
        f, J = nonlinear_system(x)

        # Αν όλες οι τιμές του f(x) είναι κοντά στο 0, έχουμε λύση
        if np.all(np.abs(f) < tol):
            break

        # Λύνει το J·δx = f για να βρει διόρθωση
        dx = gauss_solve(J, f)
        x -= dx  # επόμενη προσέγγιση

    print("Λύση:", x)

# Εκτέλεση
main()
```

```python
# Υπολογισμός ιδιοτιμών 3x3 πίνακα χωρίς χρήση βιβλιοθηκών

# Συνάρτηση υπολογισμού ριζών κυβικής εξίσωσης με τον τύπο του Cardano
def solve_cubic(a, b, c):
    # Εξίσωση: x^3 + a*x^2 + b*x + c = 0
    p = b - a**2 / 3
    q = (2 * a**3) / 27 - (a * b) / 3 + c
    
    D = (q / 2)**2 + (p / 3)**3

    roots = []

    if D > 0:
        # μία πραγματική ρίζα
        sqrt_D = (q/2)**0.5
        u = (-q/2 + sqrt_D)**(1/3)
        v = (-q/2 - sqrt_D)**(1/3)
        root = u + v - a / 3
        roots.append(root)

    elif D == 0:
        u = (-q/2)**(1/3)
        root1 = 2*u - a/3
        root2 = -u - a/3
        roots.extend([root1, root2])

    else:
        import math
        r = (-p**3 / 27)**0.5
        phi = math.acos(-q / (2 * r))
        r = (-p / 3)**0.5
        for k in range(3):
            theta = (phi + 2 * math.pi * k) / 3
            root = 2 * r * math.cos(theta) - a / 3
            roots.append(root)

    return roots

# Ορισμός πίνακα A 3x3
A = [
    [6.3, 2.1, 4.15],
    [3.1, 5.14, 1.03],
    [-11.0, 12.3, -8.8]
]

# Υπολογισμός χαρακτηριστικού πολυωνύμου: x^3 + a*x^2 + b*x + c = 0
trace = A[0][0] + A[1][1] + A[2][2]

# Tr(A^2)
trA2 = (
    A[0][0]**2 + A[1][1]**2 + A[2][2]**2
    + 2*A[0][1]*A[1][0] + 2*A[0][2]*A[2][0] + 2*A[1][2]*A[2][1]
)

b = 0.5 * (trace**2 - trA2)

det = (
    A[0][0]*A[1][1]*A[2][2] + A[0][1]*A[1][2]*A[2][0] + A[0][2]*A[1][0]*A[2][1]
    - A[0][2]*A[1][1]*A[2][0] - A[0][0]*A[1][2]*A[2][1] - A[0][1]*A[1][0]*A[2][2]
)

# Συντελεστές πολυωνύμου
coeff_a = -trace
coeff_b = b
coeff_c = -det

# Λύση εξίσωσης
eigenvalues = solve_cubic(coeff_a, coeff_b, coeff_c)

# Εκτύπωση ιδιοτιμών
for val in eigenvalues:
    print(val, 0.0)  # Μόνο πραγματικές ρίζες (η φανταστική = 0)
```

```python
# 1β # NEWTON - diko mou - tsekara leitourgei ok
import numpy as np

filename = 'C:/Users/souli/Downloads/points.txt' #allagi onomatos filename apo edw

# Υπολογίζει το γινόμενο (x - x_0)(x - x_1)...(x - x_{i-1})
def compute_basis_product(i, x_values, eval_point):
    if i == 0:
        return 1.0
    else:
        return compute_basis_product(i - 1, x_values, eval_point) * (eval_point - x_values[i - 1])

# Υπολογίζει τους συντελεστές του πολυωνύμου παρεμβολής στη μορφή Newton
def compute_newton_coefficients(x_values, y_values):
    n = len(x_values)
    coefficients = np.zeros(n)

    coefficients[0] = y_values[0]

    for j in range(1, n):
        coefficients[j] = y_values[j]
        for i in range(j):
            coefficients[j] -= coefficients[i] * compute_basis_product(i, x_values, x_values[j])
        coefficients[j] /= compute_basis_product(j, x_values, x_values[j])

    return coefficients

# Υπολογίζει την τιμή του πολυωνύμου παρεμβολής Newton σε ένα σημείο
def evaluate_newton_polynomial(x_values, y_values, eval_point):
    coefficients = compute_newton_coefficients(x_values, y_values)
    result = 0.0
    for i in range(len(x_values)):
        result += coefficients[i] * compute_basis_product(i, x_values, eval_point)
    return result

# Διαβάζει σημεία (x, y) από αρχείο
def load_data_from_file(filename):
    with open(filename, "r") as file:
        n_points = int(file.readline())
        x_vals = np.zeros(n_points)
        y_vals = np.zeros(n_points)
        for i in range(n_points):
            line = file.readline().split()
            x_vals[i] = float(line[0])
            y_vals[i] = float(line[1])
    return x_vals, y_vals

# Κύρια συνάρτηση
def main():
    x_data, y_data = load_data_from_file(filename)
    x_start = x_data.min()
    x_end = x_data.max()

    total_points = 100
    step = (x_end - x_start) / (total_points - 1)

    # Εκτύπωση τίτλων στηλών
    print(f"{'x':>10} {'P(x)':>20} {'sin(x)':>20}")
    print("-" * 70)

    for i in range(total_points):
        x_val = x_start + i * step
        interpolated = evaluate_newton_polynomial(x_data, y_data, x_val)
        exact = np.sin(x_val)
        error = abs(interpolated - exact)
        print(f"{x_val:10.6f} {interpolated:20.6f} {exact:20.6f} ")

main()
```

```python
# ASKISI 2  diko mou OK alla vgazei nan...
import numpy as np

# Κανονικοποίηση για αριθμητική σταθερότητα
def scale_rows(matrix, rhs, start_index):
    n = len(rhs)
    for i in range(start_index, n):
        max_val = abs(rhs[i])
        for j in range(start_index, n):
            max_val = max(max_val, abs(matrix[i, j]))
        for j in range(start_index, n):
            matrix[i, j] /= max_val
        rhs[i] /= max_val

# Εναλλαγή γραμμών
def swap_rows(matrix, rhs, row1, row2):
    n = len(rhs)
    for j in range(row1, n):
        matrix[row1, j], matrix[row2, j] = matrix[row2, j], matrix[row1, j]
    rhs[row1], rhs[row2] = rhs[row2], rhs[row1]

# Pivoting με μέγιστο στοιχείο κατά στήλη
def pivot(matrix, rhs, current_step):
    n = len(rhs)
    scale_rows(matrix, rhs, current_step)
    max_row = current_step
    for i in range(current_step + 1, n):
        if abs(matrix[i, current_step]) > abs(matrix[max_row, current_step]):
            max_row = i
    if max_row != current_step:
        swap_rows(matrix, rhs, current_step, max_row)

# Τριγωνοποίηση
def forward_elimination(matrix, rhs):
    n = len(rhs)
    for k in range(n - 1):
        pivot(matrix, rhs, k)
        for i in range(k + 1, n):
            multiplier = -matrix[i, k] / matrix[k, k]
            for j in range(k, n):
                matrix[i, j] += multiplier * matrix[k, j]
            rhs[i] += multiplier * rhs[k]

# Οπισθοδρόμηση
def back_substitution(matrix, rhs):
    n = len(rhs)
    solution = np.zeros(n)
    for i in range(n - 1, -1, -1):
        solution[i] = rhs[i]
        for j in range(i + 1, n):
            solution[i] -= matrix[i, j] * solution[j]
        solution[i] /= matrix[i, i]
    return solution

# Συνδυαστική επίλυση συστήματος με Gauss
def solve_system(matrix, rhs):
    forward_elimination(matrix, rhs)
    return back_substitution(matrix, rhs)

# Πολυωνυμική αξιολόγηση
def evaluate_polynomial(x, coefficients):
    return sum(coeff * x**i for i, coeff in enumerate(coefficients))

# Δημιουργία γραμμικού συστήματος για ρητή παρεμβολή
def construct_rational_system(x_data, y_data, numerator_degree, denominator_degree):
    n = len(x_data)
    matrix = np.zeros((n, n))
    rhs = y_data.copy()

    for i in range(n):
        for j in range(numerator_degree + 1):
            matrix[i, j] = x_data[i]**j
        for j in range(1, denominator_degree + 1):
            matrix[i, numerator_degree + j] = -y_data[i] * x_data[i]**j

    return matrix, rhs

# Επίλυση και αξιολόγηση ρητής παρεμβολής στο x
def evaluate_rational(x, x_data, y_data, numerator_degree, denominator_degree):
    matrix, rhs = construct_rational_system(x_data, y_data, numerator_degree, denominator_degree)
    solution = solve_system(matrix, rhs)

    num_coeffs = solution[:numerator_degree + 1]
    denom_coeffs = np.insert(solution[numerator_degree + 1:], 0, 1.0)  # q_0 = 1

    numerator = evaluate_polynomial(x, num_coeffs)
    denominator = evaluate_polynomial(x, denom_coeffs)

    return numerator / denominator

# Διαβάζει σημεία από αρχείο
def read_data(filename):
    with open(filename, "r") as f:
        n = int(f.readline())
        x_vals = np.zeros(n)
        y_vals = np.zeros(n)
        for i in range(n):
            x, y = f.readline().split()
            x_vals[i] = float(x)
            y_vals[i] = float(y)
    return x_vals, y_vals

# Κύρια ροή προγράμματος
def main():
    x_data, y_data = read_data(filename)
    M = 2  # βαθμός αριθμητή
    N = 3  # βαθμός παρονομαστή

    # Είσοδος χρήστη
    x_input = float(input("Δώσε μια τιμή x για παρεμβολή: "))
    interpolated = evaluate_rational(x_input, x_data, y_data, M, N)
    print(f"\nΤιμή ρητής παρεμβολής στο x = {x_input:.6f} είναι: {interpolated:.6f}\n")

    # Σύγκριση ρητής παρεμβολής με αρχικά y δεδομένα
    print(f"{'x':>10} {'y (αρχική)':>20} {'y (παρεμβολή)':>25} ")
    print("-" * 75)
    for xi, yi in zip(x_data, y_data):
        ri = evaluate_rational(xi, x_data, y_data, M, N)
        error = abs(yi - ri)
        print(f"{xi:10.6f} {yi:20.6f} {ri:25.6f} ")

main()
```

```python
# ΑΣΚΗΣΗ # ΔΙΚΟ ΜΟΘ ΤΣΕΚ SPLINE
# Κανονικοποίηση γραμμών για αριθμητική σταθερότητα
def scale_rows(matrix, rhs, start):
    n = len(rhs)
    for i in range(start, n):
        max_val = abs(rhs[i])
        for j in range(start, n):
            max_val = max(max_val, abs(matrix[i, j]))
        for j in range(start, n):
            matrix[i, j] /= max_val
        rhs[i] /= max_val

# Εναλλαγή γραμμών
def swap_rows(matrix, rhs, row1, row2):
    n = len(rhs)
    for j in range(row1, n):
        matrix[row1, j], matrix[row2, j] = matrix[row2, j], matrix[row1, j]
    rhs[row1], rhs[row2] = rhs[row2], rhs[row1]

# Pivot για σταθερότητα
def pivot(matrix, rhs, k):
    n = len(rhs)
    scale_rows(matrix, rhs, k)
    max_row = k
    for i in range(k + 1, n):
        if abs(matrix[i, k]) > abs(matrix[max_row, k]):
            max_row = i
    if max_row != k:
        swap_rows(matrix, rhs, k, max_row)

# Τριγωνοποίηση
def forward_elimination(matrix, rhs):
    n = len(rhs)
    for k in range(n - 1):
        pivot(matrix, rhs, k)
        for i in range(k + 1, n):
            factor = -matrix[i, k] / matrix[k, k]
            for j in range(k, n):
                matrix[i, j] += factor * matrix[k, j]
            rhs[i] += factor * rhs[k]

# Οπισθοδρόμηση
def back_substitution(matrix, rhs):
    n = len(rhs)
    solution = np.zeros(n)
    for i in range(n - 1, -1, -1):
        solution[i] = rhs[i]
        for j in range(i + 1, n):
            solution[i] -= matrix[i, j] * solution[j]
        solution[i] /= matrix[i, i]
    return solution

# Συνδυασμός
def solve_linear_system(matrix, rhs):
    forward_elimination(matrix, rhs)
    return back_substitution(matrix, rhs)

# Δημιουργία και επίλυση συστήματος spline
def compute_spline_coefficients(x_vals, y_vals):
    n = len(x_vals) - 1
    total_eqs = 3 * n
    A = np.zeros((total_eqs, total_eqs))
    B = np.zeros(total_eqs)

    # Υπολογισμός των εξισώσεων
    row = 0
    for i in range(n):
        dx = x_vals[i + 1] - x_vals[i]
        A[row, i] = dx**2
        A[row, i + n] = dx
        A[row, i + 2 * n] = 1.0
        B[row] = (y_vals[i + 1] - y_vals[i]) / dx
        row += 1

    for i in range(n - 1):
        dx = x_vals[i + 1] - x_vals[i]
        A[row, i] = 3 * dx**2
        A[row, i + n] = 2 * dx
        A[row, i + 2 * n] = 1.0
        A[row, i + 2 * n + 1] = -1.0
        row += 1

    for i in range(n - 1):
        dx = x_vals[i + 1] - x_vals[i]
        A[row, i] = 3 * dx
        A[row, i + n] = 1.0
        A[row, i + n + 1] = -1.0
        row += 1

    # Οριακές συνθήκες
    A[row, n] = 1.0
    row += 1
    A[row, n - 1] = 3 * (x_vals[n] - x_vals[n - 1])
    A[row, 2 * n - 1] = 1.0

    # Επίλυση συστήματος
    X = solve_linear_system(A, B)
    a = X[:n]
    b = X[n:2 * n]
    c = X[2 * n:]
    d = y_vals[:n]

    return a, b, c, d

# Αξιολόγηση spline στο σημείο z
def evaluate_spline(a, b, c, d, x_vals, z):
    n = len(x_vals) - 1
    i = 0
    while i < n and x_vals[i + 1] < z:
        i += 1
    if z > x_vals[n]:
        i = n - 1
    dx = z - x_vals[i]
    return ((a[i] * dx + b[i]) * dx + c[i]) * dx + d[i]

# Ανάγνωση αρχείου (x, y)
def read_data(filename):
    with open(filename, "r") as file:
        n = int(file.readline())
        x = np.zeros(n)
        y = np.zeros(n)
        for i in range(n):
            x_val, y_val = file.readline().split()
            x[i] = float(x_val)
            y[i] = float(y_val)
    return x, y

# Κύρια συνάρτηση
def main():
    x_vals, y_vals = read_data(filename)
    a, b, c, d = compute_spline_coefficients(x_vals, y_vals)

    x_min, x_max = x_vals.min(), x_vals.max()
    steps = 100
    dx = (x_max - x_min) / (steps - 1)

    # Εκτύπωση αποτελεσμάτων
    print(f"{'x':>10} {'Spline(x)':>20} {'sin(x)':>20} {'|σφάλμα|':>20}")
    print("-" * 70)

    for i in range(steps):
        x_eval = x_min + i * dx
        spline_val = evaluate_spline(a, b, c, d, x_vals, x_eval)
        exact_val = np.sin(x_eval)
        error = abs(spline_val - exact_val)
        print(f"{x_eval:10.6f} {spline_val:20.6f} {exact_val:20.6f} {error:20.6f}")

main()

```
